#### import important libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


```

```python
data=pd.read_csv("Iris.csv")
data.head()
```

```python
df=data
```

#### Importing libraries for Decision tree classifier and Splitting the dataset

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
model=tree.DecisionTreeClassifier(criterion='entropy',splitter='best')
```

```python
x=df.iloc[:,1:5]
y=df['Species']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

#### Training the dataset and testing it.

```python
run=model.fit(X_train,Y_train)
Y_pred=run.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred))
```

#### Checking the accuracy

```python
 print(accuracy_score(Y_test,Y_pred))
```

```python
print(confusion_matrix(Y_test,Y_pred))
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(Y_pred,Y_test),annot=True)
```

```python
col_names=list(df.columns.values)
```

#### Plotting the decision tree

```python
plt.figure(figsize=(25,13))
tree.plot_tree(model.fit(x,y),feature_names=col_names,filled=True,precision=3,proportion=True,rounded=True)
plt.show()
```
